Addition of vectors

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])

sum=vector_1+vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='g')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,5,1, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-6,6)
plt.ylim(-6,6)
plt.show()
```

Subtraction of vectors

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])

subtract=vector_1-vector_2
print(subtract)
```

Dot product of vectors

```python
a=np.asarray([2,3])

b=np.asarray([4,4])

a_dot_b=np.dot(a,b)

print(a_dot_b)
```

```python
c=np.asarray([40,20,35])

d=np.asarray([53,24,68])

c_dot_d=np.dot(c,d)

print(c_dot_d)
```

Cross product of vectors

```python
a=np.asarray([2,3])

b=np.asarray([4,4])

a_cross_b=np.cross(a,b)

print(a_cross_b)
```

```python
c=np.asarray([40,20,35])

d=np.asarray([53,24,68])

c_cross_d=np.cross(c,d)

print(c_cross_d)
```

Projetion of 'a' vector on 'v' vector

```python
a=np.array([5,7])

v=np.array([2,-8])

#magnitude of vector v
magnitude_v=np.sqrt(sum(v**2))

a_dot_v=np.dot(a,v)

projection_of_a_on_v=(a_dot_v/magnitude_v**2)*v

print(projection_of_a_on_v)
```
